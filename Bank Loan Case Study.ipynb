```python
def Uni_Analysis_Numarical(dataframe, column): 
    sns.set(style='darkgrid') 
    plt.figure(figsize=(25, 5)) 
 
    plt.subplot(1, 3, 1) 
    sns.boxplot(data=dataframe, x=column, orient='v').set(title='Box Plot') 
    plt.subplot(1, 3, 2) 
    sns.distplot(dataframe[column].dropna()).set(title='Distplot')  
    plt.show()
```

```python
def Uni_Analysis_Categorcal(dataframe, column): 
    sns.set(style='darkgrid') 
    plt.figure(figsize = [12,5]) 
    dataframe[column].value_counts().plot.barh(width = 0.8) 
    plt.title(column) 
    plt.show() 
```

```python
for i in Numerical_Data: 
    Uni_Analysis_Numarical(df,i)
```

```python
Categorical_Data_1 = ['NAME_CONTRACT_TYPE','FLAG_OWN_CAR','FLAG_OWN_REALTY','NAME_TYPE_SUITE','NAME_INCOME_TYPE','NAME_EDUCATION_TYPE','NAME_FAMILY_STATUS','NAME_HOUSING_TYPE','OCCUPATION_TYPE','AGE_IN_YEARS_RANGE','EMPLOYMENT_YEARS_RANGE','AMT_CREDIT_in_lakhs_Range','AMT_INCOME_TOTAL_RANGE'] 
 
Numarical_Data_1 = ['AMT_ANNUITY','AMT_GOODS_PRICE','CNT_FAM_MEMBERS','CNT_CHILDREN','Credit_Ratio'] 
 
Tagget_Variable_Payment_Difficulty = df[df.TARGET == 1] 
 
Tagget_Variable_All_Other = df[df.TARGET == 0] 
 
Tagget_Variable_All_Other.CODE_GENDER.value_counts() 
```

```python

def Tagget_categorical_Uni(variable, Tagget_Variable_Payment_Difficulty, Tagget_Variable_All_Other):
    plt.style.use('classic')
    sns.despine()

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(25, 8))

    # Plot for Payment Difficulty
    sns.countplot(x=variable, data=Tagget_Variable_Payment_Difficulty, linewidth=1, ax=ax1, edgecolor=sns.color_palette("dark", 3), hue=variable)
    ax1.set_ylabel('Total Counts')
    ax1.set_title(f'Distribution of {variable} - Payment Difficulty', fontsize=18)
    ax1.set_xticklabels(ax1.get_xticklabels(), rotation=40, ha="right")

    for p in ax1.patches:
        height = p.get_height()
        ax1.annotate('{:.1f}%'.format((height / len(Tagget_Variable_Payment_Difficulty)) * 100), 
                     (p.get_x() + p.get_width() / 2., height + 100), ha='center')

    # Plot for All Other
    sns.countplot(x=variable, data=Tagget_Variable_All_Other, linewidth=1, ax=ax2, edgecolor=sns.color_palette("dark", 3), hue=variable)
    ax2.set_ylabel('Total Counts')
    ax2.set_title(f'Distribution of {variable} - All Other', fontsize=18)
    ax2.set_xticklabels(ax2.get_xticklabels(), rotation=40, ha="right")

    for p in ax2.patches:
        height = p.get_height()
        ax2.annotate('{:.1f}%'.format((height / len(Tagget_Variable_All_Other)) * 100), 
                     (p.get_x() + p.get_width() / 2., height + 100), ha='center')

    plt.show()
        
                  
```

```python
Tagget_categorical_Uni('CODE_GENDER', Tagget_Variable_Payment_Difficulty, Tagget_Variable_All_Other)
```

```python
for i in Categorical_Data_1:
    Tagget_categorical_Uni(i, Tagget_Variable_Payment_Difficulty, Tagget_Variable_All_Other)
```

```python
def Tagget_Numarical_Bi(variable_1, variable_2, Tagget_Variable_Payment_Difficulty, Tagget_Variable_All_Other):
    plt.style.use('ggplot')
    sns.despine()
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 6))
    
    # Scatter plot for Payment Difficulty
    sns.scatterplot(x=variable_1, y=variable_2, data=Tagget_Variable_Payment_Difficulty, ax=ax1)
    ax1.set_xlabel(variable_1)     
    ax1.set_ylabel(variable_2)
    ax1.set_title(f'{variable_1} vs {variable_2} - Payment Difficulty', fontsize=15)
    
    # Scatter plot for All Other
    sns.scatterplot(x=variable_1, y=variable_2, data=Tagget_Variable_All_Other, ax=ax2)
    ax2.set_xlabel(variable_1)     
    ax2.set_ylabel(variable_2)
    ax2.set_title(f'{variable_1} vs {variable_2} - All Other', fontsize=15)
    
    plt.tight_layout()  
    plt.show()
```
